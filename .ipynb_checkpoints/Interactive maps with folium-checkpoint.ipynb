#### Createing interactive maps using folium
#### Author: Mussa Ngosha, Data Scientist

```python
Folium is a Python library used for visualizing geospatial data. It is easy to use and yet a powerful library. 
Folium is a Python wrapper for Leaflet.js which is a leading open-source JavaScript library for plotting interactive maps.
```

#### Import library

```python
import folium as flm
```

#### Specify the area of interest by using locations (Latitude and Longitude)

```python
Tanzania_map = flm.Map(location = [-6.3728,34.8925], zoom_start = 6,width=550,height=500)
```

```python
Tanzania_map
```

#### Use different tiles to display an interactive maps
#### 01: Stamen Tonners to display a map when exploring river meanders and coastal zones

```python
Tanzania_map = flm.Map(location = [-6.3728,34.8925],zoom_start = 6,tiles = 'Stamen Toner',width=550,height=500)
Tanzania_map
```

#### 02: Stamen Terrain to display a map when exploring hills and natural vegetation

```python
Tanzania_map = flm.Map(location = [-6.3728,34.8925], zoom_start = 6,tiles = 'Stamen Terrain',width=550,height=500)
Tanzania_map
```

#### 03: Stamen watercolor to display a map when exploring waterbodies

```python
Tanzania_map = flm.Map(location = [-6.3728,34.8925], zoom_start = 6,tiles = 'Stamen watercolor',width=550,height=500)
Tanzania_map
```

#### 07: Open streetMap

```python
Tanzania_map = flm.Map(location = [-6.3728,34.8925], zoom_start = 6,tiles = 'OpenStreetMap',width=550,height=500)
Tanzania_map
```

#### Use different tiles on a single map

```python
flm.TileLayer('Stamen Terrain').add_to(Tanzania_map)
flm.TileLayer('Stamen Toner').add_to(Tanzania_map)
flm.TileLayer('Stamen Water Color').add_to(Tanzania_map)
flm.TileLayer('cartodbpositron').add_to(Tanzania_map)
flm.TileLayer('cartodbdark_matter').add_to(Tanzania_map)
flm.LayerControl().add_to(Tanzania_map)
Tanzania_map
```

#### Add Markers to the Map
#### Zonal distribution of Referral Hospitals in Tanzania

```python
These hospitals are Kilimanjaro Christian Medical Centre (KCMC) in the Northern zone, Bugando Medical Centre (BMC)
in the Western zone, Mbeya Referral Hospital (MRH) in the Southern highlands zone, and Muhimbili National Hospital (MNH) 
which serves the coastal zone
```

```python
flm.Marker(location=[-2.5290, 32.9057],popup='<h3 style="color:green;">BMC</h3>').add_to(Tanzania_map)
flm.Marker(location=[-3.3448, 37.3300],popup='<h3 style="color:green;">KCMC</h3>').add_to(Tanzania_map)
flm.Marker(location=[-6.8149, 37.3294],popup='<h3 style="color:green;">MNH</h3>').add_to(Tanzania_map)
flm.Marker(location=[-8.8894, 33.4387],popup='<h3 style="color:green;">MRH</h3>').add_to(Tanzania_map)
Tanzania_map
```
